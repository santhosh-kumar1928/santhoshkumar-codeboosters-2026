```python
import pandas as pd
print("Libraries imported successfully")
```

```python
df = pd.read_csv('student_performance.csv')
print(df)
```

```python
print('number of students',df.shape[0])
print('number of columns',df.shape[1])
print('number of departments',df['department'].nunique())
```

```python
print(df['department'].value_counts())
```

```python
print('Highest Math Score:',df['math_score'].max())
print('Lowest Math Score:',df['math_score'].min())
print('Average Math Score:',df['math_score'].mean())
print('Average Programmingscore:',df['programming_score'].mean())
print(f'Average Attendance: {df['attendance_percentage'].mean():.2f}%')
```

```python
df['total_score'] = df['math_score'] + df['programming_score']
print('Total score column created.')
```

```python
print("\n[4] GENDER WISE AVERAGE SCORES")
print("-"*40)
gender_scores=df.groupby('gender')['math_score'].mean().round(2)
print(gender_scores)
```

```python
print("\n[5] top 5 perfomers")
print("-"*40)
top5=df.sort_values('math_score',ascending=False).head(5)
for index,row in top5.iterrows():
  print(f"{row['name']:<20}|{row['department']:<18}|{row['math_score']}")
```

```python
print("\n[6] STUDENT WITH LOW ATTENDANCE(below 75%)")
print("-"*40)
low_attendance=df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_attendance)>0:
  print(low_attendance.to_string(index=False))

else:
  print("All students have attendance above 75%")
  print("\n"+"="*60)
  print("Report complete")
  print("="*60)
```
